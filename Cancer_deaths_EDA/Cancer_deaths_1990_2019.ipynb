# Cancer and Deaths Dataset : 1990~2019 Globally

The "Cancer and Deaths Dataset: 1990~2019 Globally" is a comprehensive dataset containing information on cancer incidence and mortality rates across the world from 1990 to 2019.
The dataset is an excellent resource for researchers, healthcare professionals, and policymakers who are interested in understanding the global burden of cancer and its impact on populations.

[ Total 9 file , 160 columns, All the countries, 30Years data ]
>In 2017, 9.6 million people are estimated to have died from the various forms of cancer. Every sixth death in the world is due to cancer, making it the second leading cause of death – second only to cardiovascular diseases.1

Progress against many other causes of deaths and demographic drivers of increasing population size, life expectancy and — particularly in higher-income countries — aging populations mean that the total number of cancer deaths continues to increase. This is a very personal topic to many: nearly everyone knows or has lost someone dear to them from this collection of diseases.

# 1. Percentage of deaths per type

```python
import pandas as pd
death_per_type= pd.read_csv('data/annual-number-of-deaths-by-cause.csv')
```

We rename the columns in order to be easier understood, the results refer to all genders and ages

```python
columns= ['Entity', 'Code', 'Year',
       'Executions',
       'Meningiti',
       'Alzheimer',
       'Parkinson',
       'Nutritional deficiencies',
       'Malaria',
       'Drowning',
       'Interpersonal violence',
       'Maternal disorders',
       'HIV/AIDS',
       'Drug use disorders',
       'Tuberculosis',
       'Cardiovascular diseases',
       'Lower respiratory infections',
       'Neonatal disorders',
       'Alcohol use disorders',
       'Self-harm',
       'Exposure to forces of nature',
       'Diarrheal diseases',
       'Environmental heat and cold exposure',
       'Neoplasms',
       'Conflict and terrorism',
       'Diabetes mellitus',
       'Chronic kidney disease',
       'Poisonings',
       'Protein-energy malnutrition',
       'Terrorism',
       'Road injuries',
       'Chronic respiratory diseases',
       'Cirrhosis and other chronic liver diseases',
       'Digestive diseases',
       'Fire, heat, and hot substances',
       'Acute hepatitis']

#death_per_type=death_per_type.set_index('Year')
d = dict(zip(death_per_type.columns,columns))

death_per_type.rename(columns=d,inplace=True)
```

```python
death_per_type_per_year=death_per_type.groupby('Year').sum().iloc[:,:].reset_index()
```

```python
# Scale the data to percentage
death_per_type_per_year['total']=death_per_type_per_year.iloc[:, 1:].sum(axis=1)
for i in range(len(death_per_type_per_year)):
    death_per_type_per_year.iloc[i,1:33] = round(death_per_type_per_year.iloc[i,1:33]*100/death_per_type_per_year.iloc[i,33],2)
death_per_type_per_year
```

The main causes of death per year (here are only displayed the death causes that have a value >= 3%)

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=death_per_type_per_year.shape[0]//2, ncols=2, figsize=(12, 5*(death_per_type_per_year.shape[0]//2)))


# Iterate through each row of the dataframe and create a pie plot for that row
for index, row in death_per_type_per_year.iterrows():
    #print(death_per_type_per_year.Year[index])
    labels = []
    values = []
    for i, val in enumerate(row[1:33]):
        if val >= 3:
            labels.append(death_per_type_per_year.columns[i+1])
            values.append(val)
            #print(death_per_type_per_year.columns[i+1],' : ', val, '%')

    axs[index//2, index%2].pie(values, labels=labels, autopct='%1.1f%%', pctdistance=0.8, labeldistance=1.1, startangle=90)
    axs[index//2, index%2].set_title(f"Year {death_per_type_per_year.Year[index]}")

plt.tight_layout()
plt.show()
```

The main causes of death seem to be the cardiovascular diseases followed by neoplasms.

A summary for 3 main causes with the percentages from 1990 to 2019:

-  **cardiovascular diseases**: 36.7% -> 44.9% 
-  **neoplasms**: 17.9% -> 25.2% 
-  **lower respiratory infections**: 9.4% -> 5.8%

Important Note:
- HIV/AIDS causes only started having a value higher than 3% in 2001 where there was a stable percentage between 2001 to 2007 of around 4%

```python
from numpy.random import default_rng
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


def dotplot(input_x, **args):

    # Count how many times does each value occur
    unique_values, counts = np.unique(input_x, return_counts=True)
    
    # Convert 1D input into 2D array
    scatter_x = [] # x values 
    scatter_y = [] # corresponding y values
    for idx, value in enumerate(unique_values):
        for counter in range(1, counts[idx]+1):
            scatter_x.append(value)
            scatter_y.append(counter)

    # draw dot plot using scatter() 
    plt.scatter(scatter_x, scatter_y, **args)
    
    # Optional - show all unique values on x-axis. 
    # Matplotlib might hide some of them  
    plt.gca().set_xticks(unique_values)

# Get random colors
rng = default_rng()
# We need colors for all data points  
# And each color consists of 3 numbers (RGB)
# Hence input size = (hs_heights.size, 3) 
random_colors = rng.random(size=(age_per_country.size, 3))

# Set seaborn style and plot size
sns.set_theme(style='ticks')
plt.figure(figsize=(10, 6), dpi=150)

# draw dot plot
# Set random colors using the 'color' parameter
dotplot(input_x=age_per_country, s=200, color=random_colors)
plt.show()
```

```python
import seaborn as sns

sns.set_theme(style="dark")
flights = sns.load_dataset("flights")

# Plot each year's time series in its own facet
g = sns.relplot(
    data=flights,
    x="month", y="passengers", col="year", hue="year",
    kind="line", palette="crest", linewidth=4, zorder=5,
    col_wrap=3, height=2, aspect=1.5, legend=False,
)

# Iterate over each subplot to customize further
for year, ax in g.axes_dict.items():

    # Add the title as an annotation within the plot
    ax.text(.8, .85, year, transform=ax.transAxes, fontweight="bold")

    # Plot every year's time series in the background
    sns.lineplot(
        data=flights, x="month", y="passengers", units="year",
        estimator=None, color=".7", linewidth=1, ax=ax,
    )

# Reduce the frequency of the x axis ticks
ax.set_xticks(ax.get_xticks()[::2])

# Tweak the supporting aspects of the plot
g.set_titles("")
g.set_axis_labels("", "Passengers")
g.tight_layout()
```
